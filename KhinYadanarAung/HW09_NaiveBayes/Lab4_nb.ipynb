```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/Yadanar025/Supervised-Machine-Learning/refs/heads/main/data/fraud.csv', index_col = 0)
y = df['Class'].values
df = df.iloc[:,1:]
X = df.drop(columns = 'Class').values
```

```python
x_min = min(X[2])
x_max = max(X[2])

print("Minimum of x:", x_min)
print("Maximum of x:", x_max)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                    test_size = 0.40,
                                    random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_mnb = scaler.fit_transform(X_train)
X_test_mnb = scaler.transform(X_test)

mnb = MultinomialNB(alpha=1.0)
mnb.fit(X_train_mnb, y_train)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB(alpha=1.0)
bnb.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score


ypred_test_b = bnb.predict(X_test)
mat_clf = confusion_matrix(y_test, ypred_test_b)
report_clf = classification_report(y_test, ypred_test_b)
print("Y Test_Bays")
print(mat_clf)
print(report_clf)

ypred_testb = bnb.predict_proba(X_test)
auc = roc_auc_score(y_test, ypred_testb[:,1])
print(auc)
```

```python
ypred_train_b = bnb.predict(X_train)
mat_clf = confusion_matrix(y_train, ypred_train_b)
report_clf = classification_report(y_train, ypred_train_b)
print("Y Train_Bays")
print(mat_clf)
print(report_clf)

ypred_trainb = bnb.predict_proba(X_train)
auc = roc_auc_score(y_train, ypred_trainb[:,1])
print(auc)
```

```python
ypred_test_g = gnb.predict(X_test)
mat_clf = confusion_matrix(y_test, ypred_test_g)
report_clf = classification_report(y_test, ypred_test_g)
print("Y Test_Gussian")
print(mat_clf)
print(report_clf)

ypred_testg = gnb.predict_proba(X_test)
auc = roc_auc_score(y_test, ypred_testg[:,1])
print(auc)
print("Y Train_Gussian")
ypred_train = gnb.predict(X_train)
mat_clf = confusion_matrix(y_train, ypred_train)
report_clf = classification_report(y_train, ypred_train)

print(mat_clf)
print(report_clf)

ypred_traing = gnb.predict_proba(X_train)
auc = roc_auc_score(y_train, ypred_traing[:,1])
print(auc)
```

```python
ypred_test_m = mnb.predict(X_test)
mat_clf = confusion_matrix(y_test, ypred_test_m)
report_clf = classification_report(y_test, ypred_test_m)
print("Y Test_Multinomial")
print(mat_clf)
print(report_clf)

ypred_testm = mnb.predict_proba(X_test)
auc = roc_auc_score(y_test, ypred_testm[:,1])
print(auc)
print("Y Train_Multinomial")
ypred_train_m = mnb.predict(X_train)
mat_clf = confusion_matrix(y_train, ypred_train_m)
report_clf = classification_report(y_train, ypred_train_m)

print(mat_clf)
print(report_clf)

ypred_trainm = mnb.predict_proba(X_train)
auc = roc_auc_score(y_train, ypred_trainm[:,1])
print(auc)
```
